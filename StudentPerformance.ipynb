# Does studied_hours increase performance_score?

# Result
#Key Finding:
There is a positive relationship between Study Hours and Performance Score. Students who spend more time studying tend to achieve higher performance scores.
#Method Used:
- Data cleaning
- Correlation analysis
- Scatter plot with regression line (Seaborn)
- Data visualization using Python
#Insight:
Study time plays an important role in academic success, but consistency and balance are also important factors.
#Tools Used:
Python | Pandas | Seaborn | Matplotlib


```python
plt.figure(figsize=(12,6),facecolor='darkgray')
sns.set_style("whitegrid")

sns.scatterplot(
    x="Study Hours",
    y="Performance Score",
    data=df,
    hue='Extracurricular Activities',
    palette="colorblind",
    alpha =0.6,
    s=100
)
sns.regplot(
    data= df,
    x="Study Hours",
    y="Performance Score",
    scatter= False,
    color= 'red',
    line_kws={'linewidth': 2}
)
plt.title("Study Hours Vs Performance",fontweight = 'bold',fontsize = 14)
plt.xlabel('Study Hours',fontsize=14, fontweight ='bold')
plt.ylabel('Performance', fontsize = 14, fontweight ='bold')

plt.tight_layout()
plt.show()
plt.close()
```

# Does Sleep Hours affect Performance Score?

# final conclusion
The correlation between sleep hours and performance index is 0.048, indicating almost no linear relationship. This suggests that sleep duration alone does not significantly impact student performance in this dataset.

# Analysis
Analysis shows that sleep hours have negligible correlation (r ≈ 0.05) with student performance, indicating that other factors play a more important role.

# Does practicing sample papers improve performance?

#  Does previous_score strongly affect final performance?
The correlation coefficient between Previous Score and Performance Score is approximately 0.92, indicating a strong positive linear relationship. This suggests that prior academic performance is a significant predictor of current performance outcomes.

📊 Data Insight:
One of the strongest patterns I found in my analysis was the relationship between previous performance and current results. With a correlation of r ≈ 0.92, the data shows a very strong positive relationship.

# Do students with extracurricular activities perform better or worse?

```python
df['Extracurricular Activities'].corr(df['Performance Score'])
```

```python
plt.figure(figsize=(12,6),facecolor='darkgray')
sns.set_style("whitegrid")

ax = sns.boxenplot(
    data=df,
    x='Extracurricular Activities',
    y ='Performance Score',
    hue='Extracurricular Activities',
    palette='colorblind',
    legend= False
)


plt.title("Extracurricular Activities Vs Performance",fontweight = 'bold',fontsize = 14)
plt.xlabel('Extracurricular Activities',fontsize=14, fontweight ='bold')
plt.ylabel('Performance', fontsize = 14, fontweight ='bold')


#handles, labels = ax.get_legend_handles_labels()
#ax.legend(handles, ["No","Yes"])
plt.tight_layout()
plt.show()

```

# final
The boxplot shows that students with and without extracurricular activities have very similar performance distributions. The median, spread, and range are nearly identical, indicating that extracurricular activities do not have a significant impact on performance in this dataset.

- Extracurricular activities show minimal impact on performance, as both groups exhibit nearly identical distributions.

# Which factor affects performance the most?

```python
plt.figure(figsize=(12,6),facecolor='darkgray')
sns.set_style("whitegrid")

sns.heatmap(
    data=df.corr(),
    annot= True,
    cmap='coolwarm'
)

#plt.xticks(rotation=30)
#plt.yticks(rotation=30)
plt.title("“Correlation between academic and lifestyle factors", fontweight="bold", fontsize = 14)
plt.tight_layout()
plt.show()
```

📌 Data Insight: Correlation Heatmap Analysis

I analyzed the relationships between different academic factors using a correlation heatmap.

The most notable finding is the very strong positive correlation between Previous Score and Performance Score (r ≈ 0.92), indicating that past academic performance is the strongest predictor of current performance.

On the other hand, variables such as Sleep Hours (r ≈ 0.05) and Practice Papers (r ≈ 0.04) show almost no meaningful correlation with performance, suggesting that quantity-based effort alone does not strongly influence outcomes in this dataset.

📊 Overall insight:

Strongest driver of performance → Previous academic results
Weak or negligible impact → Sleep hours, practice papers

💡 Key takeaway:
Success is more strongly linked to consistent past performance than isolated effort metrics, highlighting the importance of long-term learning patterns.
